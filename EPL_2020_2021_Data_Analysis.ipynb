```python
# Import the Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset

epl_df = pd.read_excel('/content/drive/MyDrive/Python Projects/Football Data Analysis Project/EPL_20_21.xlsx')
epl_df.head()
```

```python
# Getting Information About the Dataset 

epl_df.info()
```

```python
# Describe Statistical Data

epl_df.describe()
```

```python
# Counting "na" Values If Exists

epl_df.isna().sum()
```

```python
# Create 2 New Columns

epl_df['Average_Mins_Per_Match'] = (epl_df['Mins'] / epl_df['Matches']).astype(np.int64)
epl_df['Average_Goals_Per_Match'] = (epl_df['Goals'] / epl_df['Matches']).astype(np.float64)
epl_df.head()
```

```python
# Calculating Total Goals

total_goals = epl_df['Goals'].sum()
print('Number of Total Goals:',total_goals)
```

```python
# Calculating Total Penalty Goals

total_penalty_goals = epl_df['Penalty_Goals'].sum()
print('Number of Total Penalty Goals:', total_penalty_goals)
```

```python
# Total Penalty Attempts

total_penalty_attempts = epl_df['Penalty_Attempted'].sum()
print('Number of Total Penalty Attempts:', total_penalty_attempts)
```

```python
# Pie Chart for Penaties Missed vs Scored

plt.figure(figsize=(15,6))
penalty_not_scored = total_penalty_attempts - total_penalty_goals

# Pie Chart Parameters
data = [penalty_not_scored, total_penalty_goals]
labels = ['Penalties Missed', 'Penalties Scored']

# For More Information About Color Palette, Please Visit:
# https://seaborn.pydata.org/generated/seaborn.color_palette.html
color = sns.color_palette('Set2')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

```python
# Finding Unique Positions

epl_df['Position'].unique()
```

```python
# Total FW Players
fw_players = epl_df[epl_df['Position'] == 'FW']
fw_players
```

```python
total_fw_players = np.size(fw_players, axis=0)
print('Total Number of Forward Players:', total_fw_players)
```

```python
# Number of Different Nations that Players Come From
total_nationality = np.size(epl_df['Nationality'].unique())
print(' Number of Different Nations that Players Come From:', total_nationality)
```

```python
# Number of Players From Which Countries

nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind='bar', figsize=(12,6), color=sns.color_palette('magma'))
```

```python
# Clubs with Maximum Players in Their Squad

epl_df['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
# Clubs with Least Players in Their Squad

epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette('viridis'))
```

```python
# Playera Based on the Age Group

age_under_20 = epl_df[epl_df['Age'] <= 20]
age_20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age_25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
age_above_30 = epl_df[epl_df['Age'] > 30]
```

```python
# Showing Pie Chart Each Age Group Percentage

x = np.array([age_under_20['Name'].count(),age_20_25['Name'].count(),age_25_30['Name'].count(),
age_above_30['Name'].count()])

labels = ["<= 20", "> 20 & <=25", "> 25 & <=30","> 30"]
plt.title('Total Player with Age Groups', fontsize=20)
plt.pie(x, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
# Total Number of Players Each Team Age Under 20

age_under_20['Club'].value_counts().plot(kind='bar', color=sns.color_palette('cubehelix'))
```

```python
# U20 Players in Manchester United

age_under_20[age_under_20['Club'] == "Manchester United"]
```

```python
# Box and Whisker Plots Each Club on Ages

plt.figure(figsize=(12,6))
sns.boxplot(x='Club', y='Age', data=epl_df)
plt.xticks(rotation=90)
```

```python
# Mean Age of All Players Group by Clubs

age_mean = epl_df.groupby('Club')['Age'].mean().sort_values(ascending=False)
age_mean
```

```python
# Total Assists of Each Club

assists_of_clubs = epl_df.groupby('Club', as_index=False)['Assists'].sum()

sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=assists_of_clubs.sort_values(by='Assists'), palette='Set2')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of Clubs VS Total Assists', fontsize=20)
```

```python
# Top 10 Assists

top_10_assists = epl_df[['Name', 'Club', 'Assists', 'Matches']].nlargest(10, columns='Assists')
top_10_assists.index = np.arange(1, len(top_10_assists) + 1)
top_10_assists
```

```python
goals_by_clubs = epl_df.groupby('Club', as_index=False)['Goals'].sum()
ax = sns.barplot(x='Club', y='Goals', data=goals_by_clubs.sort_values(by='Goals'), palette='rocket')
ax.set_xlabel('Club', fontsize=30)
ax.set_ylabel('Goals', fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of Clubs VS Total Goals', fontsize=20)
```

```python
# Top 10 Goals

top_10_goals = epl_df[['Name', 'Club', 'Goals', 'Matches']].nlargest(10, columns='Goals')
top_10_goals.index = np.arange(1, len(top_10_goals) + 1)
top_10_goals
```

```python
top_10_goals_per_match = epl_df[['Name', 'Average_Goals_Per_Match', 'Goals', 'Matches']].nlargest(10, columns='Average_Goals_Per_Match')
top_10_goals_per_match.index = np.arange(1,len(top_10_goals_per_match) + 1)
top_10_goals_per_match
```

```python
# Creating Pie Chart for Goals with Assists and without Assists

plt.figure(figsize=(14,7))
assists = epl_df['Assists'].sum()
data = [total_goals - assists, assists]
labels = ["Goals without Assists","Goals with Assits"]
color = sns.color_palette('Set1')
plt.pie(data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

```python
# Top 10 Players with Most Yellow Cards

epl_yellow_card = epl_df.sort_values(by='Yellow_Cards', ascending=False)[:10]

plt.figure(figsize=(20,6))
plt.title('Top 10 Yellow Cards')
c = sns.barplot(x=epl_yellow_card['Name'], y=epl_yellow_card['Yellow_Cards'], label='Players', color='yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(), rotation=45)
c
```
